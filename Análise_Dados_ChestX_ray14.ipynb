
# Texto Resumo sobre DenseNet-121
Ocorrem problemas quando as CNNs precisam de mais camadas. Isso se deve pois o caminho das informações da camada de entrada até a camada de saída (e para o gradiente na direção oposta) torna-se tão grande que as informações podem desaparecer antes de chegar ao outro lado.

Os autores resolvem o problema garantindo o máximo fluxo de informação (e gradiente). Para fazer isso, eles simplesmente conectam todas as camadas diretamente umas às outras.
Em vez de extrair poder de representação de arquiteturas extremamente profundas ou amplas, o DenseNets explora o potencial da rede por meio da reutilização de recursos.

DenseNets não somam os features de saída da camada com os features de entrada da camada, mas os concatena.

Adaptado da Fonte: https://towardsdatascience.com/understanding-and-visualizing-densenets-7f688092391a



---


"Dentre as razões que levaram à escolha desta arquitetura estão o conhecimento das principais vantagens da DenseNet em relação as outas redes, que são a redução do desaparacimento do gradiente (Gradient Vanishing), o fortalecimento da propagação das características, o incentivo a reutilização e a redução drástica de parâmetros." 

"A DenseNet requer aprender menos parâmetros que outras CNN pois não há a necessidade de aprender mapas de carac- terísticas redundantes. A arquitetura feed-forward podem ser vistas como um algoritmo com um estado, que é passado de camada em camada, gravando o estado anterior para a camada subsequente, alterando este estado e preservando informações que precisam ser preservadas"

"A DenseNet faz uma diferenciação explícita do que é informação adicionada e preservada, e man- tém as suas camadas com saídas pequenas (12 mapas gerados por camada, por exemplo), adicionando este pequeno grupo de novos mapas ao “conhecimento coletivo” existente. Ao final o classificador toma sua decisão baseado nos mapas ger- ados em toda a rede. Outra vantagem é o fluxo aprimorado de informações e gradientes pela rede, o que facilita o treina- mento. Cada camada tem acesso ao gradiente pela função de perda e pelo sinal original, o que ajuda a treinar redes mais profundas, além disso, as conexões densas geram um efeito regularizador, reduzindo o ajuste excessivo em conjuntos de treino menores. A concatenação de mapas de caraterística de diferentes camadas aumenta a variedade de características nas entradas e melhora a eficiência, sendo uma das principais diferenças entre a DenseNet e a ResNet, por exemplo."

"Cada camada convolucional gera um número fixo de mapas de características que é concatenado ao conjunto de mapas recebidos das camadas anteriores, fazendo com que a cada camada sejam criados novos mapas que exploram caracterís- ticas mais profundas. Este valor que define a quantidade de mapas de características gerados a cada camada é chamado de taxa de crescimento ou valor k."

"Os blocos são camadas abstratas que encapsulam sub-camadas que realizam o processo de convolução, extração de característica e subamostragem. Existem dois tipos de blocos para serem descritos, os blocos densos e os blocos de transição."

"O bloco denso é a camada que gera os mapas de características definido pelo valor k, ele possui como entrada os mapas provenientes das camadas anteriores e dentro dele há camadas densas que geram os novos mapas a serem concate- nados. Porém o autor trata como camada densa a sequência de três (ou quatro) operações: a normalização em lote dos mapas de entrada, a transformação via função de ativação ReLU nos mapas normalizados e por fim uma convolução com filtro 3 × 3 sem a utilização do Bias e mantendo o mesmo padding."

"Ainda existe uma versão da DenseNet que insere uma camada densa extra no início das outras, mas com uma convolução 1 × 1 que **gera quatro vezes o valor de k mapas de característica** e auxilie a encontrar características ainda mais profundas, esta camada densa extra é chamada de camada de gargalo, ou bottleneck e quando utilizada decidiu denominar a rede como DenseNet-B"

"...vantagens da DenseNet em relação as outas redes, que são a redução do desaparacimento do gradiente (Gradient Vanishing), o fortalecimento da propagação das características, o incentivo a reutilização e a redução drástica de parâmetros."

Figura acima representa um bloco denso com 5 camadas (L=5). Cada camada está conectada com todas as suas subsequentes. Portanto a quantidade de conexões é de L*(L+1)/2 = 15. Diferentemente de uma rede tradicional em que cada camada está ligada a próxima. No exemplo acima, em uma rede tradicional, teríamos apenas 5

**Layers** no KERAS: São usados para construir os modelos

**Tensors** no KERAS: Usados para definir o fluxos de dados através do modelo

Ao contrário de Resnets, a DenseNet (Dense Convolutional Network) não combina recursos por somatório, mas combina os recursos concatenando-os. Portanto, a camada 'i' tem 'i' entradas e consiste em mapas de características de todos os seus blocos convolucionais anteriores. Isso introduz conexões ‘(I (I + 1)) / 2’ na rede, em vez de apenas conexões ‘I’ como nas arquiteturas deep learning tradicionais. Portanto, requer menos parâmetros do que as redes neurais convolucionais tradicionais, pois não há necessidade de aprender mapas de características sem importância.

https://towardsdatascience.com/creating-densenet-121-with-tensorflow-edbc08a956d8

"A rede DenseNet, desenvolvida por (HUANG et al., 2017), representa o próximo passo no desenvolvimento de redes convolucionais ainda mais profundas do que as ResNets."

"...quanto maior a profundidade da rede convolucional, maior é o caminho para que a informação contida na camada de entrada flua para a camada de saída, fazendo com o que os gradientes, que fluem em direção oposta, sumam quando estiverem calculando os pesos em camadas iniciais."

"A DenseNet foi desenvolvida tendo como característica principal a criação de caminhos curtos que permitem o fluxo máximo de informação entre as camadas na rede, conectando todas as camadas (com mesmas as dimensões de mapas de características) diretamente cada uma com as outras."

"Para aprimorar ainda mais o fluxo de informações na rede, utiliza-se a concatenação das camadas anteriores ao invés de somatório"

"Semelhantemente à ResNet, esta concatenação de mapas de características devem ter dimensões iguais dentro de um mesmo bloco chamado de bloco denso. Entre dois blocos densos tem a camada de transição, responsável pelas diminuição do tamanho dos mapas de características. Nas camadas de transição, ocorrem uma convolução 1 × 1 e uma operação de pooling."

http://www.ele.ufes.br/sites/engenhariaeletrica.ufes.br/files/field/anexo/projeto_de_graduacao_ii_-_matheus_coutinho_cavalcantes.pdf

Um filtro 1 × 1 pode ser usado para criar uma projeção dos mapas de características.

O número de mapas de características criados será o mesmo e o efeito pode ser um refinamento das características já extraídas. Isso geralmente é chamado de channel-wise pooling, em oposição ao pool tradicional por caractísticas em cada canal.

https://machinelearningmastery.com/introduction-to-1x1-convolutions-to-reduce-the-complexity-of-convolutional-neural-networks/

## **Importar imagens com base no DataFrame do Pandas gerado**

```
def importar_imagens(train_df,valid_df):
  train_gen = core_idg.flow_from_dataframe(dataframe=train_df, 
                              directory=None,
                              x_col = 'path',
                              y_col = 'newLabel', 
                              class_mode = 'categorical',
                              classes = all_labels,
                              target_size = IMG_SIZE,
                              color_mode = 'grayscale',
                              batch_size = 32,
  )

  valid_gen = core_idg.flow_from_dataframe(dataframe=valid_df, 
                              directory=None,
                              x_col = 'path',
                              y_col = 'newLabel', 
                              class_mode = 'categorical',
                              classes = all_labels,
                              target_size = IMG_SIZE,
                              color_mode = 'grayscale',
                              batch_size = 256,
  )
  test_X, test_Y = next(core_idg.flow_from_dataframe(dataframe=valid_df, 
                              directory=None,
                              x_col = 'path',
                              y_col = 'newLabel', 
                              class_mode = 'categorical',
                              classes = all_labels,
                              target_size = IMG_SIZE,
                              color_mode = 'grayscale',
                              batch_size = 1024,
  ))

  return train_gen, valid_gen, test_X, test_Y
```

```
def ResNet50_parameter(shape):
  base_model = tf.keras.applications.ResNet50(input_shape = shape, 
                                  include_top = False,
                                  weights=None,
                                  )
  base_model.trainable = False
  multi_disease_model = Sequential()
  multi_disease_model.add(base_model)
  multi_disease_model.add(Flatten())
  multi_disease_model.add(Dense(14,activation='softmax'))
  multi_disease_model.compile(optimizer = 'adam', loss = 'binary_crossentropy',
                            metrics = ['binary_accuracy', 'mae'])
  return multi_disease_model
```
